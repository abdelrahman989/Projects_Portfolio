```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

## Data cleaning

```python
df.isnull().sum().reset_index()
```

```python

df = df.dropna()
```

```python
df.isnull().sum().reset_index()
```

```python
df.dtypes
```

```python
#checking for outliers
df.boxplot(column=['gross'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
#sorting our values by gross profitt
df = df.sort_values('gross',ascending=False)
```

```python
df.head()
```

```python
sns.pairplot(df)          
```

## Scatter Plot (Budget Vs Gross)

```python
plt.style.use('ggplot')
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'bold',
        'size': 20,
        }
x=df['budget']
y=df['gross']
plt.figure(figsize=(15,10),dpi=300)
plt.title('Budget Vs Gross',fontdict=font)
plt.xlabel('Budget',fontdict=font)
plt.ylabel('Gross',fontdict=font)
plt.scatter(x,y)
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10),dpi=300)
sns.regplot(x,y,scatter_kws={'color':'black'},line_kws={'color':'blue'})
```

```python
df.corr(method ='pearson')
```

```python
df.corr(method ='kendall')
```

```python
df.corr(method ='spearman')
```

```python
correlation_matrix = df.corr()
plt.figure(dpi=300)
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_number = df.copy()
```

```python
for col in df_number.columns:
    if df_number[col].dtype == 'object' :
        df_number[col]=df_number[col].astype('category')
        df_number[col] = df_number[col].cat.codes
```

```python
df_number.head()
```

```python
df.head()
```

```python
correlation_matrix = df_number.corr()
plt.figure(dpi=300)
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
